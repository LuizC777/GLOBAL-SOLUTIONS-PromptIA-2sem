```python
import openai
from google.colab import userdata

client = openai.OpenAI(api_key=userdata.get('OPENAI_API_KEY'))

def chatbot_aca(usuario, mensagem_usuario):
    # Definir o papel do assistente
    if usuario == 1:
        sistema = (
            "Peça para o usuário fornecer o trabalho atual dele e suas principais tarefas para estimar o risco de automação e identificar as habilidades mais críticas para o futuro daquela área."
        )
    elif usuario == 2:
        sistema = (
            "Com base na profissão e nas habilidades críticas, sugira 3 a 5 áreas de aprimoramento e, para cada uma, sugerir um recurso de aprendizado "
        )
    elif usuario == 3:
        sistema = (
            "Faça uma sugestão de Reskilling ao usuário, para que possa se realocar no mercado de trabalho. Mapeie as habilidades transferíveis e sugira 3 a 5 habilidades novas."
        )
    elif usuario == 4:
        sistema = (
            "Faça uma entrevista com o usuário com três perguntas para uma nova função, considerando a clareza relenvacia e profundidade."
        )

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": sistema},
            {"role": "user", "content": mensagem_usuario}
        ],
        temperature=0.3,
        max_tokens=600
    )

    # Acesso do conteúdo do objeto de resposta de v1
    resposta = response.choices[0].message.content
    return resposta

print("Agente de Carreira Adaptativa")
print("1. Análise de Perfil e Risco de Automação.")
print("2. Plano de Upskilling Personalizado.")
print("3. Sugestão de Caminho de Reskilling")
print("4. Simulação de Entrevista \n")

entry = int(input("Escolha uma opção: "))

while True:
    user_input = input("\nVocê: ")

    if user_input.lower() in ["sair", "exit", "quit"]:
        print("Encerrando o agente.")
        break

    resposta = chatbot_aca(entry, user_input)
    print(f"\n Agente: {resposta}")
```
